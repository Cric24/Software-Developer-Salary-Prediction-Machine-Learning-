```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey_results_public.csv")

df.head()
```

```python
df = df[["Country", "ConvertedComp"]]
df = df.rename({"ConvertedComp": "Salary"}, axis=1)
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 108)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 250000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()

```

```python
X = df[['Country']]
y = df['Salary']
```

```python
from sklearn.tree import DecisionTreeRegressor

dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y)
```

```python
import pickle
```

```python
data = {"model": dec_tree_reg, "le_country": le_country}
with open('saved_steps.pkl', 'wb') as file:
    pickle.dump(data, file)
```

```python
with open('saved_steps.pkl', 'rb') as file:
    data = pickle.load(file)

    
regressor_loaded = data["model"]
le_country = data["le_country"]

```

```python
y_pred = regressor_loaded.predict(X)
y_pred
```

```python
import numpy as np
X_sample = np.array([le_country.transform(['United States'])[0]])  # Note the removal of an extra set of square brackets
X_sample = X_sample.astype(float)
y_pred_sample = regressor_loaded.predict(X_sample.reshape(1, -1))  # Reshape to ensure a 2D array
print("${:,.02f}".format(y_pred_sample[0]))

```
